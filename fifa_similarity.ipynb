```python
import pandas as pd

players = pd.read_csv("players_22.csv")
players
```

```python
features = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
info = ['sofifa_id', 'short_name', 'nationality_name', 'age', 'player_positions', 'overall', 'value_eur']
columns = info + features 

df = players.loc[:, columns]
df.set_index('sofifa_id', inplace=True) 
df.dropna(inplace=True)

df
```

```python
stats = df[features]
stats
```

```python
df_range = (stats.max(axis=0) - stats.min(axis=0))
ndf = (stats - stats.min(axis=0)) / df_range
ndf
```

```python
from sklearn.metrics.pairwise import cosine_distances, euclidean_distances

def find_similar_players(player_id, df, feature_df, sim_metric):
    player = feature_df.loc[player_id]
    dist = sim_metric(feature_df, [player])[:, 0]
    query_distances = list(zip(feature_df.index, dist))
    
    for similar_player_id, similarity_score in sorted(query_distances, key=lambda x: x[1])[:11]:
        similar_player = df.loc[similar_player_id, 'short_name']
        print(similar_player_id, similar_player, similarity_score)

query_ids = [158023, 20801, 190871] # Messi, Ronaldo, Neymar
```

```python
print("Cosine Distances\n")
for query_id in query_ids:
    print(f"Top 10 similar players for {df.loc[query_id, 'short_name']}:")
    find_similar_players(query_id, df, ndf, cosine_distances)
    print("\n")

print("Euclidean Distances\n")
for query_id in query_ids:
    print(f"Top 10 similar players for {df.loc[query_id, 'short_name']}:")
    find_similar_players(query_id, df, ndf, euclidean_distances)
    print("\n")
```
